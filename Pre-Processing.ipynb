```python
from pandas import read_csv
import scipy 
import matplotlib.pyplot as plt
import datetime
import warnings 
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
import six
import joblib
import sys
sys.modules['sklearn.externals.six'] = six
sys.modules['sklearn.externals.joblib'] = joblib
from pmdarima import auto_arima
from scipy import io
import pandas as pd
```

```python
#read matlab file
mat_file = scipy.io.loadmat('InsulinGlucoseData2.mat')
mat_file = {i:j for i,j in mat_file.items() if i[0] != '_'}

mat_file
```

```python
#cgm data
cgm_data=pd.DataFrame()
cgm_data['dateNumber']=pd.Series(mat_file['dateNumber'][0])
cgm_data['numCGM']=list(mat_file['numCGM'][0])
cgm_data.to_csv("CGMData.csv")

cgm_data
```

```python
#bolus data
bolus_data=pd.DataFrame()
bolus_data['dateMuBolus']=pd.Series(mat_file['dateMuBolus'][0])
bolus_data['actBolusDelivered']=list(mat_file['actBolusDelivered'][0])
bolus_data.to_csv("BolusData.csv")

bolus_data
```
